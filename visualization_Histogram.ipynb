```python
# Importar librerias y complementos
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
# Importar los datos al DataFrame desde el archivo CSV
df = pd.read_csv("Desercion_Clean.csv")
```

```python
# Conteo de filas y columnas
df.shape
```

```python
# Conteo de Valores unicos columna DESERTORES_AÑO_ACTUAL
df.DESERTORES_AÑO_ACTUAL.value_counts().sort_index()
```

```python
# Definir la Variable
valores_desercion = df.DESERTORES_AÑO_ACTUAL.unique()
# Definir eje X
plt.xticks(valores_desercion)
# Reemplazar coma en los valores de "miles" por punto.
formatter = FuncFormatter(lambda x, pos: f'{int(x):,}'.replace(',', '.'))
# Definir formato eje Y
plt.gca().yaxis.set_major_formatter(formatter)
# Crear titulo del histograma
plt.title('Histograma de Desertores 2024')
# Crear etiqueta ejes X y Y
plt.xlabel('Cantidad de Desertores')
plt.ylabel('Frecuencia')
# Formato letra y rotacion eje X
plt.xticks(fontsize=8,rotation=90)
# Definir bins y formato de bins
n, bins, patches = plt.hist(df.DESERTORES_AÑO_ACTUAL, color="#008c95",range=(1,50))
for i in range(len(n)):
    plt.text(bins[i] + 0.5, n[i] + 0.1, f'{int(n[i]):,}'.replace(',', '.'), ha='center', va='bottom', fontsize=10)
plt.show()
```
