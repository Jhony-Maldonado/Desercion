```python
# Importar librerias y complementos
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importar los datos al DataFrame desde el archivo CSV
df = pd.read_csv("Desercion_Clean.csv")
```

```python
# Conteo de filas y columnas
df.shape
```

```python
# Definir datos para el grafico de torta
Tabla_Modalidad = df.groupby('MODALIDAD_FORMACION')['TOTAL_APRENDICES_MATRICULADOS'].sum()
plt.title('MATRICULADOS POR MODALIDAD')
plt.pie(Tabla_Modalidad, labels=Tabla_Modalidad.index, autopct='%1.1f%%', startangle=10, colors=['#86de95', '#2596be', 'red'])
plt.legend()
```

```python
# Definir datos para el grafico de torta 2
Tabla_Modalidad = df.groupby('MODALIDAD_FORMACION')['DESERTORES_AÑO_ACTUAL'].sum()
plt.title('DESERTORES POR MODALIDAD')
plt.pie(Tabla_Modalidad, labels=Tabla_Modalidad.index, autopct='%1.1f%%', startangle=45, colors=['#86de95', '#2596be', 'red'])
plt.legend()
```
