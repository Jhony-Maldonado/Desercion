```python
# Importar librerias y complementos
import pandas as pd
import seaborn as sns
```

```python
# Importar los datos al DataFrame desde el archivo CSV
df = pd.read_csv("Desercion_Clean.csv")
```

```python
# Crear una tabla con Regionales y Cantidad de Desertores
df.groupby('NOM_REGIONAL')['DESERTORES_AÑO_ACTUAL'].sum().reset_index()
```

```python
# Crear codigo para elaborar Grafico de Barras
sns.barplot(data=df, x="DESERTORES_AÑO_ACTUAL", y="NOM_REGIONAL", estimator=sum)
```

```python
# Crear una tabla con Regionales y Cantidad de Matriculados
df.groupby('NOM_REGIONAL')['TOTAL_APRENDICES_MATRICULADOS'].sum().reset_index()
```

```python
# Crear codigo para elaborar Grafico de Barras
sns.barplot(data=df, x="TOTAL_APRENDICES_MATRICULADOS", y="NOM_REGIONAL", estimator=sum)
```
